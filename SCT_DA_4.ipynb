```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep=';')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Age'] = 2026 - df['Year_Birth']

df[['Year_Birth','Age']].head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Income'], bins=20)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

```python
df['Response'].value_counts().plot(kind='bar')
plt.title('Campaign Response')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
df['Education'].value_counts().plot(kind='bar')
plt.title('Education Distribution')
plt.xlabel('Education')
plt.ylabel('Count')
plt.show()
```

```python
df['Marital_Status'].value_counts().plot(kind='bar')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

```python
df['Response'].value_counts()
```

```python
df['Response'].value_counts().plot(kind='bar')

plt.title('Customer Response to Campaign')
plt.xlabel('Response')
plt.ylabel('Count')

plt.show()
```

```python
spending = df[['MntWines',
               'MntFruits',
               'MntMeatProducts',
               'MntFishProducts',
               'MntSweetProducts',
               'MntGoldProds']].sum()

spending.plot(kind='bar')

plt.title('Product Category Spending')
plt.ylabel('Amount Spent')

plt.show()
```

```python
channels = df[['NumWebPurchases',
               'NumCatalogPurchases',
               'NumStorePurchases']].sum()

channels.plot(kind='bar')

plt.title('Purchases by Channel')
plt.ylabel('Total Purchases')

plt.show()
```
